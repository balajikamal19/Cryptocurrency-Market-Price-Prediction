```python
from sklearn.preprocessing import LabelEncoder
var_mod = ['slug', 'symbol', 'name', 'date', 'ranknow', 'open', 'high', 'low',
       'close', 'volume', 'market_price', 'close_ratio', 'spread']
le = LabelEncoder()
for i in var_mod:
    df[i] = le.fit_transform(df[i]).astype(str)
```
